## Question 1

Floor Division and Modulo operations cannot be performed on a complex number.
Addition, Subtraction, Multiplication, Division and power operators can be performed on complex Numbers.

```python
import cmath
x = complex(5, 2)
y = complex(3, 11)
print("Addition:", x+y)
print("Subtraction:", x-y)
print("Multiplication:", x*y)
print("Division:", x/y)
print("Power:", x**y)
```

## Question 2

range() function is used to create a sequence.
It's Parameters are :
1. Starting Value
2. Ending value
3. Step
Where Start and Step are optional.

```python
for i in range(0,100,2): #creating a sequence 0 to 100 with step 2: 0,2,4,6.....100
    for j in range(99,0,-3): #creating a sequence 99 to 0 with step -3: 99,96,93.90......0
        n = i+j
        if n%6 == 0 : print(n, ": ",list(k/2 for k in range(n)))
        
```
